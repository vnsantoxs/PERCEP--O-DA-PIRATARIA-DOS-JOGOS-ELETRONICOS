```python
def contar_respostas(respostas):
    return respostas.value_counts()

def adicionar_valores(barras):
    for barra in barras:
        altura = barra.get_height()
        ax.text(barra.get_x() + barra.get_width()/2., altura + 0.1,
                f'{int(altura)}',
                ha='center', va='bottom')
        
def criar_grafico_pizza(dados, labels, titulo):
    dados_filtrados = [dado for dado in dados if dado > 0]
    labels_filtrados = [label for i, label in enumerate(labels) if dados[i] > 0]
    
    plt.figure(figsize=(6, 6))
    plt.pie(dados_filtrados, labels=labels_filtrados, autopct='%1.1f%%', startangle=140)
    plt.title(titulo)
    plt.axis('equal')
    #plt.legend(wedges,contagem.index, title=f"{titulo}", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
```
